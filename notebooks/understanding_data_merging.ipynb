MERGING DATASETS TO GET A COMBINED DATA FOR ANALYSIS (DEPENDENT VARIABLE BEING WEEKLY SALES)

```python

# Merge the datasets using appropriate keys (columns) where they are common
# Merge train_data with stores_data based on 'Store'
combined_data = data_train.merge(data_stores, on='Store', how='left')

# Merge combined_data with features_data based on 'Store' and 'Date'
combined_data = combined_data.merge(data_features, on=['Store', 'Date'], how='left')
display(combined_data.head())
# Save the DataFrame to a CSV file
combined_data.to_csv(r"..\data\raw\merged.csv", index=False) 
```
